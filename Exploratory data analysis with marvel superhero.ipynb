# Exploratory Data Analysis OF Marvel Characters

```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
marvel = pd.read_csv("charcters_stats.csv")
marvel
```

# Analysing Total Point Of All Characters

```python
marvel_total=marvel.sort_values(by=['Total'],ascending=False)
marvel_total
```

```python
plt.figure(figsize=(10,6))
plt.title("Total power of all characters")
plt.xlabel("Character  name")
plt.ylabel("Total")


plt.bar(list(marvel_total['Name'])[0:7],list(marvel_total['Total'])[0:7],color="g")
plt.show()
```

```python
marvel['Alignment'].value_counts()
```
